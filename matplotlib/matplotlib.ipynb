## Anatomy of a Figure

- figure: `fig, _ = plt.subplots()`, `fig = plt.figure()`
  - patch: `fig.patch.set()`
  - suptitle: `fig.suptitle()`, `fig.get_suptitle()`, `plt.suptitle()`
  - axes: `_, ax = plt.subplots()`, `ax = fig.add_subplot()`
    - title: `ax.set_title()`, `plt.title()`
    - spines: `ax.spines`
    - grid: `ax.grid()`, `plt.grid()`
    - lines: `ax.plot()`, `plt.plot()`
    - markers: `ax.scatter()`, `plt.scatter()`
    - legend: `ax.legend()`, `plt.legend()`
    - text: `ax.text()`, `plt.text()`
    - artist: `ax.add_artist()`
    - xaxis: `ax.xaxis`
      - label: `ax.set_xlabel()`, `plt.xlabel()`
      - xlim: `ax.set_xlim()`, `plt.xlim()`
      - major tick: `ax.set_xticks()`, `ax.tick_params(axis="x", which="major")`
        - formatter: `ax.xaxis.set_major_formatter()`
        - locator: `ax.xaxis.set_major_locator()`
      - minor tick: `ax.tick_params(axis="x", which="minor")`
        - formatter: `ax.xaxis.set_minor_formatter()`
        - locator: `ax.xaxis.set_minor_locator()`
    - yaxis: `ax.yaxis`
      - label: `ax.set_ylabel()`, `plt.ylabel()`
      - ylim: `ax.set_ylim()`, `plt.ylim()`
      - major tick: `ax.set_yticks()`, `ax.tick_params(axis="y", which="major")`
        - formatter: `ax.yaxis.set_major_formatter()`
        - locator: `ax.yaxis.set_major_locator()`
      - minor tick: `ax.tick_params(axis="y", which="minor")`
        - formatter: `ax.yaxis.set_minor_formatter()`
        - locator: `ax.yaxis.set_minor_locator()`


```python
fig, ax = plt.subplots(
    1,  # nrows
    1,  # ncols
    figsize=plt.rcParams["figure.figsize"],  # [6.4, 4.8]
    dpi=plt.rcParams["figure.dpi"],  # 100.0
    facecolor=plt.rcParams["figure.facecolor"],  # white
    edgecolor=plt.rcParams["figure.edgecolor"],  # white
    frameon=plt.rcParams["figure.frameon"],  # True
    linewidth=0.0,
    layout=None,
)
```

```python
fig, ax = plt.subplots()
fig.suptitle("Suptitle")
fig.patch.set(edgecolor="0.6", facecolor="0.9", lw=4, ls="--")
ax.set(title="Title", xlabel="X Label", ylabel="Y Label", facecolor="0.9")
```

```python
from matplotlib.ticker import AutoMinorLocator, MultipleLocator

fig, ax = plt.subplots()

ax.set(xlim=(0, 4), ylim=(0, 4))

ax.tick_params(which="major", width=1.0, length=10, labelsize=14)
ax.tick_params(which="minor", width=1.0, length=5, labelsize=10, labelcolor="0.25")

ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(1))

ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))

ax.xaxis.set_minor_formatter("{x:.2f}")
ax.yaxis.set_minor_formatter("{x:.2f}")

ax.grid(axis="both", which="minor", ls="--")
ax.grid(axis="both", which="major", ls="-")
```
