## Scalar Activations


```python
import torch
import torch.nn as nn
import torch.nn.functional as F

import matplotlib.pyplot as plt

%matplotlib inline

def show_plot(x, y, title):
    plt.figure(figsize=(8,4))
    plt.title(title)
    plt.plot(x, y)
```

### ELU

Applies the Exponential Linear Unit (ELU) function, element-wise.

Method described in the paper: [Fast and Accurate Deep Network Learning by Exponential Linear Units (ELUs)](https://arxiv.org/abs/1511.07289).

ELU is defined as:

$$
        \text{ELU}(x) = \begin{cases}
        x, & \text{ if } x > 0\\
        \alpha (\exp(x) - 1), & \text{ if } x \leq 0
        \end{cases}
$$


```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.ELU.html
"""Define Layer"""

"""1. Position Arguments"""

alpha, inplace = 1.0, False

elu = nn.ELU(alpha, inplace)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

assert torch.allclose(elu(x), F.elu(x, alpha, inplace), atol=1e-6)

show_plot(x, elu(x), str(elu))
```

### Hardshrink

Applies the Hard Shrinkage (Hardshrink) function element-wise.

Hardshrink is defined as:

$$
        \text{HardShrink}(x) =
        \begin{cases}
        x, & \text{ if } x > \lambda \\
        x, & \text{ if } x < -\lambda \\
        0, & \text{ otherwise }
        \end{cases}
$$


```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.Hardshrink.html
"""Define Layer"""

"""1. Position Arguments"""

lambd = 0.5

hardshrink = nn.Hardshrink(lambd)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

assert torch.allclose(hardshrink(x), F.hardshrink(x, lambd), atol=1e-6)

show_plot(x, hardshrink(x), str(hardshrink))
```

```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.LeakyReLU.html
"""Define Layer"""

"""1. Position Arguments"""

negative_slope, inplace = 0.01, False

leaky_relu = nn.LeakyReLU(negative_slope, inplace)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

assert torch.allclose(
    leaky_relu(x), F.leaky_relu(x, negative_slope, inplace), atol=1e-6
)

show_plot(x, leaky_relu(x), str(leaky_relu))
```

### PReLU

Applies the element-wise PReLU function. Channel dim is the 2nd dim of input. When input has dims < 2, then there is no channel dim and the number of channels = 1.

$$
        \text{PReLU}(x) =
        \begin{cases}
        x, & \text{ if } x \ge 0 \\
        ax, & \text{ otherwise }
        \end{cases}
$$


```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.PReLU.html
"""Define Layer"""
"""1. Position Arguments"""

num_parameters, init = 1, 0.25

p_relu = nn.PReLU(num_parameters, init, dtype=None, device=None)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

assert torch.allclose(p_relu(x), F.prelu(x, p_relu.weight), atol=1e-6)

show_plot(x, p_relu(x).detach(), str(p_relu))
```

### RReLU

Applies the randomized leaky rectified linear unit function, element-wise.

Method described in the paper: [Empirical Evaluation of Rectified Activations in Convolutional Network](https://arxiv.org/abs/1505.00853).

The function is defined as:

$$
        \text{RReLU}(x) =
        \begin{cases}
            x & \text{if } x \geq 0 \\
            ax & \text{ otherwise }
        \end{cases}
$$

where $a$ is randomly sampled from uniform distribution $\mathcal{U}(\text{lower}, \text{upper})$ during training while during evaluation $a$ is fixed with $a = \frac{\text{lower} + \text{upper}}{2}$.


```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.RReLU.html
"""Define Layer"""
"""1. Position Arguments"""

lower, upper, inplace = 1 / 8, 1 / 3, False

r_relu = nn.RReLU(lower, upper, inplace)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

# F.rrelu(x, lower, upper, inplace)
show_plot(x, r_relu(x), str(r_relu))
```

### SELU

Applies the SELU function element-wise. More details can be found in the paper [Self-Normalizing Neural Networks](https://arxiv.org/abs/1706.02515).

$$
        \text{SELU}(x) = \text{scale} * (\max(0,x) + \min(0, \alpha * (\exp(x) - 1)))
$$

with $\alpha = 1.6732632423543772848170429916717$ and $\text{scale} = 1.0507009873554804934193349852946$.


```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.SELU.html
"""Define Layer"""
"""1. Position Arguments"""

inplace = False

selu = nn.SELU(inplace)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

assert torch.allclose(selu(x), F.selu(x, inplace), atol=1e-6)

show_plot(x, selu(x), str(selu))
```

### CELU

Applies the CELU function element-wise.

$$
        \text{CELU}(x) = \max(0,x) + \min(0, \alpha * (\exp(x/\alpha) - 1))

More details can be found in the paper [Continuously Differentiable Exponential Linear Units](https://arxiv.org/abs/1704.07483).


```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.CELU.html
"""Define Layer"""
"""1. Position Arguments"""

alpha, inplace = 1.0, False

celu = nn.CELU(alpha, inplace)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

assert torch.allclose(celu(x), F.celu(x, alpha, inplace), atol=1e-6)

show_plot(x, celu(x), str(celu))
```

### GELU

Applies the Gaussian Error Linear Units function.

$$
\text{GELU}(x) = x * \Phi(x)
$$

where $\Phi(x)$ is the Cumulative Distribution Function for Gaussian Distribution.

When the approximate argument is `'tanh'`, Gelu is estimated with:

$$
\text{GELU}(x) = 0.5 * x * (1 + \text{Tanh}(\sqrt{2 / \pi} * (x + 0.044715 * x^3)))
$$


```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.GELU.html
"""Define Layer"""
"""1. Keyword Arguments"""

approximate = "none"

gelu = nn.GELU(approximate=approximate)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

assert torch.allclose(gelu(x), F.gelu(x, approximate=approximate), atol=1e-6)

show_plot(x, gelu(x), str(gelu))
```

### Softplus

Applies the Softplus function element-wise.

$$
        \text{Softplus}(x) = \frac{1}{\beta} * \log(1 + \exp(\beta * x))
$$

SoftPlus is a smooth approximation to the ReLU function and can be used to constrain the output of a machine to always be positive.

For numerical stability the implementation reverts to the linear function when $input \times \beta > threshold$.


```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.Softplus.html
"""Define Layer"""
"""1. Position Arguments"""

beta, threshold = 1.0, 20.0

softplus = nn.Softplus(beta, threshold)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

assert torch.allclose(softplus(x), F.softplus(x, beta, threshold), atol=1e-6)

show_plot(x, softplus(x), str(softplus))
```

### Threshold

Thresholds each element of the input Tensor.

Threshold is defined as:

$$
        y =
        \begin{cases}
        x, &\text{ if } x > \text{threshold} \\
        \text{value}, &\text{ otherwise }
        \end{cases}
$$


```python
# https://docs.pytorch.org/docs/stable/generated/torch.nn.Threshold.html
"""Define Layer"""
"""1. Position Arguments"""

threshold, value, inplace = 0, 0.5, False

threshold_ = nn.Threshold(threshold, value, inplace)

"""Forward Pass"""

"""1. Inputs"""

x = torch.linspace(-7, 7, 1000)

"""2. Outputs"""

assert torch.allclose(
    threshold_(x), F.threshold(x, threshold, value, inplace), atol=1e-6
)

show_plot(x, threshold_(x), str(threshold_))
```
